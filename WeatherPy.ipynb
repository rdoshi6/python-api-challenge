```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)


# Print the city count to confirm sufficient count
cities
```

```python
#Create a data frame of the cities
Cities_df = pd.DataFrame({'Cities':cities})

#Print log of each city 
print(Cities_df)
```

```python
# Get the weather attribute data using the openweather API
url = "http://api.openweathermap.org/data/2.5/weather?"
Temperature = []
Humidity = []
Cloudiness = []
Wind_Speed  = []
Latitude = []
Longitude = []
City_Names = []
City_ID = []


# Build query URL
for city in cities:
    query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=imperial"
    response = requests.get(query_url)
    if response.status_code == 200:
        response = response.json()
        Temperature.append(response["main"]["temp"])
        Humidity.append(response["main"]["humidity"])
        Cloudiness.append(response["clouds"]["all"])
        Wind_Speed.append(response["wind"]["speed"])
        Latitude.append(response["coord"]["lat"])
        Longitude.append(response["coord"]["lon"])
        City_Names.append(response["name"])
        City_ID.append(response["id"])
        
```

```python
# Get weather attribute data in a pandas dataframe
Weather_Attribute = pd.DataFrame({"City": City_Names,"City ID": City_ID,"Temperature (F)": Temperature,"Humidity (%)": Humidity,
                                  "Cloud Coverage (%)": Cloudiness,"Wind Speed (mph)": Wind_Speed,
                                  "Latitude": Latitude, "Longitude": Longitude})

Weather_Attribute.head()
```

```python
# Export data to csv
Weather_Attribute.to_csv("output_data_file.csv")
```

```python
#Linear Regression Northern Hemisphere - Temperature (F) vs. Latitude
Latitude = Northern_Hemisphere['Latitude']
Temperature = Northern_Hemisphere['Temperature (F)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(Latitude, Temperature)
regress_values = Latitude * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(Latitude,Temperature)
plt.plot(Latitude,regress_values,"red")
plt.annotate(line_eq,(5,10),fontsize=15,color="red")

plt.ylabel("Temperature (F)")
plt.xlabel("Latitude")
plt.show()

plt.savefig("Northern Hemisphere - Temperature (F) vs. Latitude.png", dpi=100)

#After each pair of plots explain what the linear regression is modeling such as any relationships 
#you notice and any other analysis you may have.

#As you get closer to the equator, it gets warmer as expected
```
